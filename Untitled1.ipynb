```python
import ee
import geemap
import json
import pandas as pd
import matplotlib.pyplot as plt

# Initialize Earth Engine
ee.Authenticate()
ee.Initialize(project="ee-revanthbethala")

# -----------------------
# 1 Load baseline GeoJSON LineString
# -----------------------
with open("chennai.geojson") as f:
    geojson = json.load(f)
baseline_shoreline = ee.FeatureCollection(geojson)
print("Loaded baseline shoreline")

# Define AOI as a buffer around baseline
aoi = baseline_shoreline.geometry().buffer(500)  # 500 m buffer
print("Created AOI with 500m buffer")

# -----------------------
# 2 Simple Landsat collection functions
# -----------------------
def get_landsat5_collection(year, aoi):
    start = ee.Date.fromYMD(year, 1, 1)
    end = ee.Date.fromYMD(year, 12, 31)

    def process_l5(img):
        renamed = img.select(
            ['SR_B1', 'SR_B2', 'SR_B3', 'SR_B4', 'SR_B5', 'SR_B7'],
            ['blue', 'green', 'red', 'nir', 'swir1', 'swir2']
        )
        scaled = renamed.multiply(0.0000275).add(-0.2)
        ndwi = scaled.normalizedDifference(['green', 'nir']).rename('NDWI')
        return scaled.addBands(ndwi)

    if 1984 <= year <= 2012:
        return (ee.ImageCollection("LANDSAT/LT05/C02/T1_L2")
                .filterBounds(aoi)
                .filterDate(start, end)
                .filter(ee.Filter.lt('CLOUD_COVER', 50))
                .map(process_l5))
    else:
        return ee.ImageCollection([])

def get_landsat7_collection(year, aoi):
    start = ee.Date.fromYMD(year, 1, 1)
    end = ee.Date.fromYMD(year, 12, 31)

    def process_l7(img):
        renamed = img.select(
            ['SR_B1', 'SR_B2', 'SR_B3', 'SR_B4', 'SR_B5', 'SR_B7'],
            ['blue', 'green', 'red', 'nir', 'swir1', 'swir2']
        )
        scaled = renamed.multiply(0.0000275).add(-0.2)
        ndwi = scaled.normalizedDifference(['green', 'nir']).rename('NDWI')
        return scaled.addBands(ndwi)

    if year >= 1999:
        return (ee.ImageCollection("LANDSAT/LE07/C02/T1_L2")
                .filterBounds(aoi)
                .filterDate(start, end)
                .filter(ee.Filter.lt('CLOUD_COVER', 50))
                .map(process_l7))
    else:
        return ee.ImageCollection([])

def get_landsat8_collection(year, aoi):
    start = ee.Date.fromYMD(year, 1, 1)
    end = ee.Date.fromYMD(year, 12, 31)

    def process_l8(img):
        renamed = img.select(
            ['SR_B2', 'SR_B3', 'SR_B4', 'SR_B5', 'SR_B6', 'SR_B7'],
            ['blue', 'green', 'red', 'nir', 'swir1', 'swir2']
        )
        scaled = renamed.multiply(0.0000275).add(-0.2)
        ndwi = scaled.normalizedDifference(['green', 'nir']).rename('NDWI')
        return scaled.addBands(ndwi)

    if year >= 2013:
        return (ee.ImageCollection("LANDSAT/LC08/C02/T1_L2")
                .filterBounds(aoi)
                .filterDate(start, end)
                .filter(ee.Filter.lt('CLOUD_COVER', 50))
                .map(process_l8))
    else:
        return ee.ImageCollection([])

def get_landsat9_collection(year, aoi):
    start = ee.Date.fromYMD(year, 1, 1)
    end = ee.Date.fromYMD(year, 12, 31)

    def process_l9(img):
        renamed = img.select(
            ['SR_B2', 'SR_B3', 'SR_B4', 'SR_B5', 'SR_B6', 'SR_B7'],
            ['blue', 'green', 'red', 'nir', 'swir1', 'swir2']
        )
        scaled = renamed.multiply(0.0000275).add(-0.2)
        ndwi = scaled.normalizedDifference(['green', 'nir']).rename('NDWI')
        return scaled.addBands(ndwi)

    if year >= 2021:
        return (ee.ImageCollection("LANDSAT/LC09/C02/T1_L2")
                .filterBounds(aoi)
                .filterDate(start, end)
                .filter(ee.Filter.lt('CLOUD_COVER', 50))
                .map(process_l9))
    else:
        return ee.ImageCollection([])

def get_combined_collection(year, aoi):
    l5 = get_landsat5_collection(year, aoi)
    l7 = get_landsat7_collection(year, aoi)
    l8 = get_landsat8_collection(year, aoi)
    l9 = get_landsat9_collection(year, aoi)
    combined = ee.ImageCollection([]) \
        .merge(l5).merge(l7).merge(l8).merge(l9)
    return combined

# -----------------------
# 3 Extract shoreline per year
# -----------------------
def extract_shoreline(year):
    print(f" Getting Landsat collection for {year}...")
    try:
        collection = get_combined_collection(year, aoi)
        n_images = collection.size().getInfo()
        print(f" Found {n_images} images for {year}")
        if n_images == 0:
            return None, 0

        print(f" Creating composite...")
        composite = collection.median()
        ndwi = composite.select('NDWI')

        print(f" Creating water mask...")
        water_mask = ndwi.gt(-0.1)

        print(f" Detecting shoreline edges...")
        kernel = ee.Kernel.circle(radius=1)
        dilated = water_mask.focal_max(kernel=kernel, iterations=1)
        eroded = water_mask.focal_min(kernel=kernel, iterations=1)
        edges = dilated.subtract(eroded).selfMask()

        print(f" Vectorizing shoreline...")
        shoreline_vectors = edges.reduceToVectors(
            geometry=aoi,
            scale=30,
            geometryType='polygon',
            eightConnected=False,
            maxPixels=int(1e8)
        )

        def polygon_to_linestring(feature):
            coords = ee.List(feature.geometry().coordinates().get(0))
            linestring = ee.Geometry.LineString(coords)
            return ee.Feature(linestring).set('year', year)

        shoreline_lines = shoreline_vectors.map(polygon_to_linestring)
        num_shorelines = shoreline_lines.size().getInfo()
        print(f" Extracted {num_shorelines} shoreline segments")
        if num_shorelines == 0:
            return None, n_images
        return shoreline_lines, n_images

    except Exception as e:
        print(f" Error in extract_shoreline for {year}: {str(e)}")
        return None, 0

# -----------------------
# 4 Calculate distances
# -----------------------
def calculate_shoreline_distances(shoreline_fc, baseline_fc):
    try:
        baseline_geom = baseline_fc.geometry()

        def add_distance(feature):
            distance = feature.geometry().distance(baseline_geom, maxError=1)
            return feature.set('distance_m', distance)

        with_distances = shoreline_fc.map(add_distance)
        distances = with_distances.aggregate_array('distance_m')
        max_dist = distances.reduce(ee.Reducer.max()).getInfo()
        mean_dist = distances.reduce(ee.Reducer.mean()).getInfo()

        return {
            'max_distance': max_dist,
            'mean_distance': mean_dist
        }
    except Exception as e:
        print(f" Error calculating distances: {str(e)}")
        return {
            'max_distance': None,
            'mean_distance': None
        }

# -----------------------
# 5 Main analysis loop
# -----------------------
print("Starting shoreline change analysis...")
years = list(range(1990, 2025))
print(f"Processing years: {years}")
all_shorelines = ee.FeatureCollection([])
results = []

for year in years:
    print(f"\nProcessing {year}...")
    try:
        shoreline_fc, n_images = extract_shoreline(year)
        if shoreline_fc is None:
            print(f" No shoreline extracted for {year}")
            results.append({
                'year': year,
                'images_used': n_images,
                'max_dist_m': None,
                'mean_dist_m': None
            })
            continue

        else:
            print(f"No shoreline data for {year}, skipping export")
            print(f" Calculating distances to baseline...")
            distance_stats = calculate_shoreline_distances(shoreline_fc, baseline_shoreline)
        year_fc = all_shorelines.filter(ee.Filter.eq('year', year))
        count = year_fc.size().getInfo()
        if count > 0:
            filename = f"shorelines_{year}.geojson"
            filepath = os.path.join("Shoreline", filename)
            geemap.ee_export_geojson(year_fc, filename=filepath)
            print(f"Saved {count} shoreline features for {year} as {filepath}")
        results.append({
            'year': year,
            'images_used': n_images,
            'max_dist_m': distance_stats['max_distance'],
            'mean_dist_m': distance_stats['mean_distance']
        })
        if distance_stats['mean_distance'] is not None:
            print(f" Mean distance to baseline: {distance_stats['mean_distance']:.1f} m")

        all_shorelines = all_shorelines.merge(shoreline_fc)

    except Exception as e:
        print(f" Error processing {year}: {str(e)}")
        results.append({
            'year': year,
            'images_used': 0,
            'max_dist_m': None,
            'mean_dist_m': None
        })

# -----------------------
# 6 Results and Visualization
# -----------------------
df = pd.DataFrame(results)
df_valid = df.dropna(subset=['mean_dist_m']).sort_values('year')

if len(df_valid) > 0:
    df_valid['annual_change_m'] = df_valid['mean_dist_m'].diff()
    df_valid['cumulative_change_m'] = df_valid['mean_dist_m'] - df_valid['mean_dist_m'].iloc[0]
else:
    print("No valid shoreline distances available to calculate changes.")
    df_valid['annual_change_m'] = []
    df_valid['cumulative_change_m'] = []

print("\n" + "="*60)
print("SHORELINE CHANGE ANALYSIS RESULTS")
print("="*60)

# Plotting
if len(df_valid) > 0:
    print(f"\nCreating plots for {len(df_valid)} years with valid data...")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    ax1.plot(df_valid['year'], df_valid['mean_dist_m'], 'b-o', linewidth=2, markersize=6)
    if not df_valid['max_dist_m'].isna().all():
        ax1.fill_between(df_valid['year'], df_valid.get('max_dist_m', pd.Series()),
                         df_valid.get('mean_dist_m', pd.Series()), alpha=0.3, label='Range')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Distance to Baseline (m)')
    ax1.set_title('Shoreline Position Over Time')
    ax1.grid(True, alpha=0.3)
    ax1.legend()

    ax2.bar(df['year'], df['images_used'], alpha=0.7, color='skyblue')
    ax2.set_xlabel('Year')
    ax2.set_ylabel('Number of Satellite Images')
    ax2.set_title('Data Availability')
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    # plt.savefig('shoreline_analysis_plots.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("Plots saved as 'shoreline_analysis_plots.png'")

    print(f"\nSUMMARY STATISTICS:")
    print(f"Years with data: {len(df_valid)}")
    print(f"Mean distance to baseline: {df_valid['mean_dist_m'].mean():.1f} m")
    print(f"Distance range: {df_valid['mean_dist_m'].min():.1f} to {df_valid['mean_dist_m'].max():.1f} m")
    print(f"Total variation: {df_valid['mean_dist_m'].max() - df_valid['mean_dist_m'].min():.1f} m")
else:
    print("No valid data for plotting")
print(df_valid)
# Create map visualization
print(f"\nCreating map visualization...")
try:
    total_shorelines = all_shorelines.size().getInfo()
    print(f"Total shoreline segments: {total_shorelines}")
    if total_shorelines > 0:
        Map = geemap.Map(center=[13.05, 80.27], zoom=12)
        Map.addLayer(aoi, {'color': 'lightgreen', 'fillOpacity': 0.1}, 'Study Area')
        Map.addLayer(baseline_shoreline, {'color': 'blue', 'width': 3}, 'Baseline Shoreline')

        colors = ['red', 'orange', 'purple', 'cyan']
        periods = [(2000, 2009), (2010, 2019), (2020, 2025)]
        for i, (start_yr, end_yr) in enumerate(periods):
            if i < len(colors):
                period_shorelines = all_shorelines.filter(
                    ee.Filter.And(
                        ee.Filter.gte('year', start_yr),
                        ee.Filter.lte('year', end_yr)
                    )
                )
                count = period_shorelines.size().getInfo()
                if count > 0:
                    Map.addLayer(
                        period_shorelines,
                        {'color': colors[i], 'width': 2},
                        f'Shorelines {start_yr}-{end_yr} ({count})'
                    )
        print("Map visualization ready!")
        Map
    else:
        print("No shorelines to visualize")
except Exception as e:
    print(f"Error creating map: {e}")

print(f"\nAnalysis completed!")
print(f"Generated files:")
print(f" - shoreline_analysis_results.csv")
if len(df_valid) > 0:
    print(f" - shoreline_analysis_plots.png")
```

```python
def add_ndwi_l5(image):
    green = image.select('SR_B2').multiply(0.0001)
    nir = image.select('SR_B4').multiply(0.0001)
    ndwi = green.subtract(nir).divide(green.add(nir)).rename('NDWI')
    return image.addBands(ndwi)
def add_ndwi_l7(image):
    green = image.select('SR_B2').multiply(0.0001)
    nir = image.select('SR_B4').multiply(0.0001)
    ndwi = green.subtract(nir).divide(green.add(nir)).rename('NDWI')
    return image.addBands(ndwi)
def add_ndwi_l8_9(image):
    green = image.select('SR_B3').multiply(0.0001)
    nir = image.select('SR_B5').multiply(0.0001)
    ndwi = green.subtract(nir).divide(green.add(nir)).rename('NDWI')
    return image.addBands(ndwi)
def add_ndwi_s2(image):
    green = image.select('B3').multiply(0.0001)
    nir = image.select('B8').multiply(0.0001)
    ndwi = green.subtract(nir).divide(green.add(nir)).rename('NDWI')
    return image.addBands(ndwi)
```

```python
def fetch_images(year, start_date, end_date, buffer_months=2, cloud_thresh=70):
    from datetime import datetime, timedelta
    # Extend dates if needed
    start_dt = datetime.strptime(f"{year}-{start_date}", "%Y-%m-%d") - timedelta(days=buffer_months*30)
    end_dt = datetime.strptime(f"{year}-{end_date}", "%Y-%m-%d") + timedelta(days=buffer_months*30)
    start_str, end_str = start_dt.strftime("%Y-%m-%d"), end_dt.strftime("%Y-%m-%d")
    # Collection/year selection
    if year <= 2011:
        cid, ndwi_fun, scale, cloud_key = "LANDSAT/LT05/C02/T1_L2", add_ndwi_l5, 30, 'CLOUD_COVER'
    elif 2012 <= year <= 2013:
        cid, ndwi_fun, scale, cloud_key = "LANDSAT/LE07/C02/T1_L2", add_ndwi_l7, 30, 'CLOUD_COVER'
    elif 2014 <= year <= 2021:
        cid, ndwi_fun, scale, cloud_key = "LANDSAT/LC08/C02/T1_L2", add_ndwi_l8_9, 30, 'CLOUD_COVER'
    else:
        cid, ndwi_fun, scale, cloud_key = "COPERNICUS/S2_SR", add_ndwi_s2, 10, 'CLOUDY_PIXEL_PERCENTAGE'
    collection = (ee.ImageCollection(cid)
                  .filterBounds(aoi)
                  .filterDate(start_str, end_str)
                  .filter(ee.Filter.lt(cloud_key, cloud_thresh))
                  .map(ndwi_fun)
                  .select('NDWI'))
    return collection, scale,cloud_key
```

```python
def shoreline_vector(water_mask, scale):
    return water_mask.reduceToVectors(
        geometry=aoi, scale=scale, geometryType='polygon', maxPixels=1e9)
```

```python
import geemap
years = range(1994, 2008)
season_start, season_end = "06-01", "09-30"  # You can make this a loop for seasons

Map = geemap.Map(center=[16.245, 81.246], zoom=12)
Map.addLayer(aoi, {}, "AOI")
def otsu_threshold(image, scale):
    # Compute histogram over AOI
    histogram = image.reduceRegion(
        reducer=ee.Reducer.histogram(),
        geometry=aoi,
        scale=scale,
        maxPixels=1e13
    ).get('NDWI')

    histogram = ee.Dictionary(histogram)
    counts = ee.Array(histogram.get('histogram'))
    bins = ee.Array(histogram.get('bucketMeans'))

    # Otsu algorithm
    total = counts.reduce(ee.Reducer.sum(), [0]).get([0])
    sum_total = bins.multiply(counts).reduce(ee.Reducer.sum(), [0]).get([0])
    mean_total = ee.Number(sum_total).divide(total)

    size = counts.length().get([0])
    best_thresh = ee.Number(0)
    best_between = ee.Number(0)

    for i in range(size.getInfo()):
        weight1 = counts.slice(0, 0, i).reduce(ee.Reducer.sum(), [0]).get([0])
        weight2 = total.subtract(weight1)
        mean1 = bins.slice(0, 0, i).multiply(counts.slice(0, 0, i)) \
            .reduce(ee.Reducer.sum(), [0]).get([0])
        mean2 = bins.slice(0, i).multiply(counts.slice(0, i)) \
            .reduce(ee.Reducer.sum(), [0]).get([0])

        mean1 = ee.Number(mean1).divide(weight1)
        mean2 = ee.Number(mean2).divide(weight2)
        between = ee.Number(weight1).multiply(weight2) \
            .multiply(mean1.subtract(mean2).pow(2))

        best_thresh = ee.Algorithms.If(between.gt(best_between), bins.get(i), best_thresh)
        best_between = ee.Number(ee.Algorithms.If(between.gt(best_between), between, best_between))

    return ee.Number(best_thresh)


for y in years:
    # Try normal
    coll, scale, cloud_key = fetch_images(y, season_start, season_end, buffer_months=0, cloud_thresh=30)
    count = coll.size().getInfo()

    # Try relaxed if needed
    if count < 1:
        coll, scale, cloud_key = fetch_images(y, season_start, season_end, buffer_months=2, cloud_thresh=55)
        count = coll.size().getInfo()

    if count == 0:
        print(f"{y}: No data available even after relaxation.")
        continue

    # Pick least cloudy image
    best_img = coll.reduce(ee.Reducer.percentile([20])).clip(aoi)
    best_img = best_img.rename(['NDWI'])


    # Dynamic NDWI threshold
    threshold = best_img.reduceRegion(
        reducer=ee.Reducer.percentile([50]),
        geometry=aoi,
        scale=scale,
        maxPixels=1e13
    ).get('NDWI')
    threshold_img = ee.Image.constant(threshold)
    water_mask = best_img.gt(threshold_img)

    # Extract shoreline
    shoreline_fc = shoreline_vector(water_mask, scale)
    Map.addLayer(shoreline_fc.style(color='blue', width=2), {}, f'Shoreline {y}')
    print(f"{y}: Shoreline extracted with dynamic threshold {threshold.getInfo()} from {count} images.")

    Map.addLayer(shoreline_fc.style(color='blue', width=2), {}, f'Shoreline {y}')
    print(f"{y}: Shoreline vector visualized. Images used: {count}")
    # Optional: Export vector to Google Drive (one year at a time)
    # task = ee.batch.Export.table.toDrive(
    #     collection=shoreline_fc,
    #     description=f'shoreline_{y}',
    #     folder='Shoreline_Results',
    #     fileFormat='GeoJSON'
    # )
    # task.start()
Map
```

```python
import geopandas as gpd
from shapely.geometry import LineString, Point
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import drive

# Mount Google Drive to access your files
drive.mount('/content/drive')

# Define your folder path and transects filename
data_folder = '/content/drive/MyDrive/Shoreline_Results'
transects_file = '/content/transects.geojson'  # Update this path to your transects GeoJSON file

# Load the transects GeoJSON as a GeoDataFrame
transect_gdf = gpd.read_file(transects_file)
print(f"Loaded {len(transect_gdf)} transects from {transects_file}")

# Load shoreline GeoJSON files for all years
years = range(1994, 2008)
geojson_files = [os.path.join(data_folder, f'shoreline_{year}.geojson') for year in years]
shorelines = {}
for year, path in zip(years, geojson_files):
    if os.path.exists(path):
        shorelines[year] = gpd.read_file(path)
        print(f"Loaded shoreline for year {year}")
    else:
        print(f"Warning: GeoJSON missing for year {year}")

# Check CRS compatibility
shoreline_crs = shorelines[1994].crs if 1994 in shorelines else None
print("Shoreline CRS:", shoreline_crs)
print("Transects CRS before alignment:", transect_gdf.crs)

# Reproject transects to shoreline CRS if different
if transect_gdf.crs != shoreline_crs:
    transect_gdf = transect_gdf.to_crs(shoreline_crs)
    print(f"Reprojected transects to CRS: {shoreline_crs}")

# Plot to visually check layer overlay
fig, ax = plt.subplots(figsize=(10, 10))
for year, gdf in sorted(shorelines.items()):
    gdf.plot(ax=ax, label=f'Shoreline {year}', alpha=0.5, zorder=1)
transect_gdf.plot(ax=ax, color='red', linewidth=2, label='Transects', zorder=2)
plt.legend()
plt.title("Shorelines and Transects Overlay")
plt.show()

# Function to find nearest intersection distance along a transect line
def shoreline_intersection_distance(shoreline_gdf, transect_line):
    intersections = shoreline_gdf.geometry.intersection(transect_line)
    points = []
    for inter in intersections:
        if inter.is_empty:
            continue
        if inter.geom_type == 'Point':
            points.append(inter)
        elif inter.geom_type == 'MultiPoint':
            points.extend([p for p in inter.geoms])
        elif inter.geom_type == 'GeometryCollection':
            for geom in inter.geoms:
                if geom.geom_type == 'Point':
                    points.append(geom)
    if not points:
        return np.nan
    origin = Point(transect_line.coords[0])
    distances = [origin.distance(p) for p in points]
    return min(distances)

# Choose baseline reference year (1994)
baseline_year = 1994
if baseline_year not in shorelines:
    raise FileNotFoundError(f"Baseline year {baseline_year} shoreline file not found")

# Compute NSM for each year compared to baseline
nsm_results = {}
for year, gdf in shorelines.items():
    distances_base = [shoreline_intersection_distance(shorelines[baseline_year], t) for t in transect_gdf.geometry]
    distances_year = [shoreline_intersection_distance(gdf, t) for t in transect_gdf.geometry]

    distances_base = np.array(distances_base)
    distances_year = np.array(distances_year)
    nsm = distances_year - distances_base
    nsm_results[year] = nsm

    print(f"Year: {year} NSM per transect: {nsm}")

print("Shoreline CRS example:", shorelines[1994].crs)
print("Transects CRS:", transect_gdf.crs)

# Plot NSM over years for each transect
plt.figure(figsize=(12, 6))
years_with_data = sorted(nsm_results.keys())
for i in range(len(transect_gdf)):
    transect_nsm_values = [nsm_results[year][i] for year in years_with_data]
    plt.plot(years_with_data, transect_nsm_values, marker='o', label=f'Transect {i+1}')

plt.xlabel('Year')
plt.ylabel('Net Shoreline Movement (units of coordinate system)')
plt.title('Shoreline Change (NSM) per Transect for Available Years')
plt.legend()
plt.grid(True)
plt.show()
```
